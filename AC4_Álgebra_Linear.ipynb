```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# função para plotagem dos gráficos
def plot_poligono(poligono, novo=True, cor='k'):
  """objeto é uma sequência de pontos"""
  x = [p[0] for p in poligono]
  y = [p[1] for p in poligono]

  if novo:
    plt.figure(figsize=(4,4))
  plt.plot(x, y, c=cor);
```

**RESPOSTA (menor ângulo):**
considerando o pentágono regular, **o menor ângulo $\alpha$ da rotação deverá ser de 72°**, para que que cada vértice coincida com a posição do vértice adjacente pois $360°/5 = 72°$.

```
# coordenadas do pentágono antes de centralizar no ponto (5, 5)
pentagono = np.array([[0, 0, 1],
                      [1, 0, 1],
                      [1+ (np.cos(np.deg2rad(72))), np.sin(np.deg2rad(72)), 1],
                      [1/2, np.tan(np.deg2rad(54))/2 + 0.5/np.cos(np.deg2rad(54)), 1],
                      [-(np.cos(np.deg2rad(72))), np.sin(np.deg2rad(72)), 1],
                      [0,0, 1]])
pentagono
```

```
# coordenadas do pentágono centrado no ponto (5, 5)
centro_pentagono_x = (0 + 1 + 1.309 + 0.5 -0.309) / 5
centro_pentagono_y = (0 + 0 + 0.951 + 1.539 + 0.951)/5
centro_pentagono = np.array([centro_pentagono_x, centro_pentagono_y, 1])
centro = np.array([5, 5])

deslocamento_pentagono = np.array([[1, 0, 1],
                                [0, 1, 1],
                                -centro_pentagono + np.array([5, 5, 2])])

pentagono_centrado = np.matmul(pentagono, deslocamento_pentagono)

plot_poligono(pentagono_centrado)
plt.plot([centro[0]], [centro[1]], '.r');
plt.title('Pentágono centrado no ponto (5,5) e lado de 1 unidade');
```

# Questão 3: $(2,0 \; pontos)$

Qual a área de um polígono com vértices em $\{(-2,-2), (-1,-1), (1,-1), (2,-2), (0,5)\}$? Fazendo $s_x = 2$, qual deve ser o valor de $s_y$ para termos um polígono com 6 vezes a área original? Mostre que o polígono após escalonado apresenta 6 vezes a área original.

Apresente um gráfico com esse polígono e o escalonamento.

**RESPOSTA 1 (área):** Para o cálculo da área do polígono, dividiu-se a figura em um triângulo e um trapézio. Assim: área total =** área do triângulo (2x6/2 = 6) + área do trapézio [(4+2)x1/2 = 3] = **9 unidades**

**RESPOSTA 2 (fator de aumento no eixo y):** Considerando $s_x = 2$, **o valor de $s_y$ deverá aumentar em 3 vezes ($s_y = 3$)**, para garantir um aumento total da área final do polígono em 6 vezes. Assim, após o escalonamento com  $s_x = 2$ e $s_y = 3$, o polígono inicial passará de uma área de 9 unidades para 54 unidades.

```
# coordenadas do polígono original
poligono3 = np.array([[-2, -2, 1],
                      [2, -2, 1],
                      [1, - 1, 1],
                      [0, 5, 1],
                      [-1, -1, 1],
                      [-2, -2, 1]])
```

```
# matriz de escalonamento do polígono
escala_poligono3 = np.array([[2,0,0],
                             [0,3,0],
                             [0,0,1]])
```

```
# plotagem do polígono original e seu escalonamento
poligono3_escala = np.matmul(poligono3, escala_poligono3)
plot_poligono(poligono3_escala)
plot_poligono(poligono3, novo = False, cor = 'r')
plt.title('Gráfico do polígono original e de seu  escalonamento em 6 vezes.');
```

# Questão 4: $(2,0 \; pontos)$

Encontre uma matriz de transformação que muda a escala de um círculo com $raio=1$ e centro em $(0,0)$ de modo que a área resultante seja o dobro da área do círculo original.

Apresente um gráfico do círculo original e após a aplicação da matriz de transformação.

#####**RESPOSTA (matriz transformação)**
- Considerando que a área inicial do círculo de raios 1 é igual a: $$A_i = pi.(1)^2 = pi$$
- Considerando que a área do círculo resultante é igual ao dobro do círculo inicial: $$A_f = 2.pi = 2pi$$
- Matriz de transformação:
$$rf = sqrt(2.pi/pi) = sqrt(2)$$
$$s_x = s_y = r_f/r_i$$
$$s_x = s_y = sqrt(2)/1 = sqrt(2)$$


```
# matriz de transformação, para que á área original (pi) passe a ser o dobro (2pi)
matriz_circulo = np.array([[2**(0.5), 0, 0],
                           [0, 2**(0.5), 0],
                           [0, 0, 1],
                           ])
matriz_circulo
```

```
# circulo original
t1 = np.linspace(0, 2*np.pi, 50)
r1 = 1
xc1, yc1 = 0, 0
x1 = r1 * np.cos(t1) + xc1
y1 = r1 * np.sin(t1) + yc1
circulo_original = np.array([x1, y1, np.ones(50)]).T
```

```
# circulo resultante após escalonamento
circulo_resultante = np.matmul(circulo_original, matriz_circulo)
```

```
# plotagem do gráfico
plot_poligono(circulo_original, cor = 'r')
plot_poligono(circulo_resultante, novo = False, cor = 'k')
plt.title('Gráfico do círculo original e após a aplicação da matriz de transformação.');

```

# Questão 5: $(2,0 \; pontos)$

Apresente um gráfico de tema livre composto unicamente por círculos obtidos a partir de transformações geométricas aplicadas sobre o círculo com $raio=1$ e centro em $(0,0)$.


```
# circulo inicial
ti= np.linspace(0, 2*np.pi, 50)
ri = 1
xci, yci = 0, 0
xi = ri * np.cos(ti) + xci
yi = ri * np.sin(ti) + yci
circulo_inicial = np.array([xi, yi, np.ones(50)]).T
```

```
# declaração das matrizes de transformação do círculo
ang = 30
matriz_aumento = np.array([[2, 0, 1],
                           [0, 2, 1],
                           [0, 0, 1]])

matriz_diminuicao = np.array([[0.05, 0, 1],
                              [0, 0.05, 1],
                              [0, 0, 1]])

desloc1 = np.array([[0, 1, 0],]*len(circulo_inicial))
desloc2 = np.array([[0, -1, 0],]*len(circulo_inicial))
desloc3 = np.array([[np.cos(np.deg2rad(ang)),
                           np.sin(np.deg2rad(ang)), 0],]*len(circulo_inicial))
desloc4 = np.array([[np.cos(np.deg2rad(-ang)),
                           np.sin(np.deg2rad(-ang)), 0],]*len(circulo_inicial))
desloc5 = np.array([[np.cos(np.deg2rad(ang + 120)),
                           np.sin(np.deg2rad(ang + 120)),
                      0],]*len(circulo_inicial))
desloc6 = np.array([[np.cos(np.deg2rad(-ang - 120)),
                           np.sin(np.deg2rad(-ang - 120)),
                      0],]*len(circulo_inicial))
```

```
# operações de transformação sobre o círculo inicial
circulo_aumentado = np.matmul(circulo_inicial, matriz_aumento)
circulo_diminuido = np.matmul(circulo_inicial, matriz_diminuicao)

circulo_deslocado1 = circulo_inicial + desloc1
circulo_deslocado2 = circulo_inicial + desloc2
circulo_deslocado3 = circulo_inicial + desloc3
circulo_deslocado4 = circulo_inicial + desloc4
circulo_deslocado5 = circulo_inicial + desloc5
circulo_deslocado6 = circulo_inicial + desloc6
```

```
# plotagem do gráfico
plt.figure(figsize=(8, 8))
plt.title('Gráfico de Mandala - Transformações sobre círculo com raio = 1.');
plot_poligono(circulo_inicial, novo = False, cor = 'b')

plot_poligono(circulo_aumentado, novo = False, cor = 'r')
plot_poligono(circulo_diminuido, novo = False, cor = 'k')

plot_poligono(circulo_deslocado1, novo = False,  cor = 'g')
plot_poligono(circulo_deslocado2, novo = False,  cor = 'g')
plot_poligono(circulo_deslocado3, novo = False,  cor = 'g')
plot_poligono(circulo_deslocado4, novo = False,  cor = 'g')
plot_poligono(circulo_deslocado5, novo = False,  cor = 'g')
plot_poligono(circulo_deslocado6, novo = False,  cor = 'g')
```
